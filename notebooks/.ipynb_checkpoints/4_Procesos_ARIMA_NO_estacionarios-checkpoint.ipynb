# Introducción

Hemos estudiado series temporales que no exhibieron un crecimiento a largo plazo, es decir no tienen una media constante. Sin embargo, muchas series de tiempo económicas y financieras no tienen una media constante. Los ejemplos incluyen: el PIB y el IPC de Ecuador.


# Diferenciación

La diferenciación de una serie es un concepto similar a la derivada. Básicamente, la diferenciación es el cambio de una serie $x$ de un periodo a otro, $\Delta x_t= x_t-x_{t-1}$. En este caso, la seria $x_t$ es integrada de orden 1 si luego de diferenciarla 1 vez se convierte en estacionaria.

De manera general, decimos que una serie es integrada de orden $d$ si luego de diferenciarla d veces se convierte en estacionaria. Es decir, $\Delta^d x_t=(1-L)^d x_t$

- Si $d=1$, $\Delta x_t= x_t-x_{t-1}$

- Si $d=2$, $\Delta^2 x_t= (1-L)^2x_t=(1-2L+L^2) x_t= (x_t -x_{t-1})-(x_{t-1}-x_{t-2})=\Delta x_t-\Delta x_{t-1}$, 


La diferenciación no elimina todas la no estacionariedad; solo elimina la asociada con la integración. En este contexto, pasamos del modelo $ARMA(p,q)$ al $ARIMA(p,d,q)$, donde $d$ indican las veces que debemos diferenciar las serie para que sea estacionaria. En primer lugar, necesitamos conocer cuántas veces hay que diferenciar una seria para convertirla en estacionaria. Luego, se procede con la parte $ARMA(p,q)$.

A continuación, exploraremos cómo la diferenciación afecta tres procesos no estacionarios diferentes: una caminata aleatoria, una caminata aleatoria con drift y un modelo de tendencia determinista. Veremos que la dificultad radica en determinar si la variable es estacionaria después de las diferencias "d". Las diferencias erróneas se suman a nuestro problema, por lo que no es una solución automática.

# Caminata aleatoria (random walk)

Consideramos el modelo $y_t = y_{t-1}+e_t$

## Generando una caminata aleatoria

```python
arparams = np.array([1, -1])
maparams = np.array([1])

nobs = 240

y = arma_generate_sample(arparams, maparams, nobs)
```

```python
y=pd.DataFrame(y, columns=['y'])
y.index = pd.period_range('1981-01-01', '2000-12-01', freq='M')
```

```python
y.plot(figsize=(12,8))
```

## Gráfico de ACF y PACF

```python
fig = plt.figure(figsize=(12,8))

#Plot ACF
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(y, lags=40, ax=ax1)

#Plot PACF
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(y, lags=40, ax=ax2)
```

## Estimación del modelo

```python
# Generar el modelo
model = sm.tsa.SARIMAX(y, order=(1,0,0), trend=None)

# Estimar los parametros del modelo
results = model.fit()

print(results.summary())
```

## Análisis de los residuos

```python
resid = results.resid
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = resid.plot(ax=ax)
```

```python
fig = plt.figure(figsize=(12,8))

#Plot ACF
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid, lags=40, ax=ax1)

#Plot PACF
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

## Estimación del modelo transformado

```python
dy = y['y'].diff(periods=1)
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = dy.plot(ax=ax)
```

```python
# Generar el modelo
model = sm.tsa.SARIMAX(y, order=(1,1,0), trend=None)

# Estimar los parametros del modelo
results = model.fit()

print(results.summary())
```

```python
resid1 = results.resid
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = resid1.plot(ax=ax)
```

# Modelo con tendencia determinística

Consideramos el modelo $y_t = \beta_0 + \beta_1 t +e_t$

## Generar el modelo con tendencia determinística

```python
arparams = np.array([1, -1])
maparams = np.array([1])

nobs = 240

y = arma_generate_sample(arparams, maparams, nobs)
```

```python
df=pd.DataFrame(y, columns=['y'])
df.index = pd.period_range('1981-01-01', '2000-12-01', freq='M')

df['e']=df['y'].diff(periods=1)
df['const']=1.2
df['trend']=range(0,240)
df['trend']= df['trend']*0.25

df['y']=df['const']+df['trend']+df['e']
```

```python
df
```

```python
y = df['y'].dropna()
```

## Gráfico ACF y PACF

```python
fig = plt.figure(figsize=(12,8))

#Plot ACF
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(y, lags=40, ax=ax1)

#Plot PACF
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(y, lags=40, ax=ax2)
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = y.plot(ax=ax)
```

## Extraer la tendencia de la serie

```python
# Generar el modelo
model = sm.tsa.SARIMAX(y, order=(0,0,0), trend='ct')

# Estimar los parametros del modelo
results = model.fit()

print(results.summary())
```

## Estimar el modelo sin tendencia

```python
# Serie sin tendencia
y_new = results.resid
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)

ax = y_new.plot(ax=ax)
```

# Caminata aleatoria con drift

Consideramos el modelo $y_t = \beta_0 + y_{t-1}+e_t$

## Generar la caminata aleatoria drift

```python
arparams = np.array([1, -1])
maparams = np.array([1])

nobs = 240

y = arma_generate_sample(arparams, maparams, nobs)
```

```python
df=pd.DataFrame(y, columns=['y_old'])
df.index = pd.period_range('1981-01-01', '2000-12-01', freq='M')

df['e']=df['y_old'].diff(periods=1)
df['const']=1.2
df['Ly']=df['y_old'].shift()
df['y']=df['const']+ df['Ly']+df['e'] 
```

```python
df.head()
```

```python
z = df['y'].dropna()
```

## Grafico ACF y PACF

```python
fig = plt.figure(figsize=(12,8))

#Plot ACF
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(z, lags=40, ax=ax1)

#Plot PACF
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(z, lags=40, ax=ax2)
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = z.plot(ax=ax)
```

## Estimar el modelo

```python
# Generar el modelo
model = sm.tsa.SARIMAX(z, order=(1,0,0), trend='c')

# Estimar los parametros del modelo
results = model.fit()

print(results.summary())
```

```python
# Serie sin tendencia
z_new = results.resid
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)

ax = z_new.plot(ax=ax)
```

## Estimar modelo transformado

```python
#  Generar el modelo
model = sm.tsa.SARIMAX(z, order=(0,1,0), trend='c')

# Estimar los parametros del modelo
results = model.fit()

print(results.summary())
```

## Ejemplo: PIB

```python
# Definir varibles
dpib = df1['pib'].diff(periods=1)
pib = df1['pib']
```
